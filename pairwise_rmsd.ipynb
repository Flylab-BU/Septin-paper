```python
import MDAnalysis as mda
from MDAnalysis.analysis import diffusionmap, align
import matplotlib.pyplot as plt
import numpy as np
print(np.__version__)


u = mda.Universe("/path/to/the/psf/file",
                 "/path/to/the/dcd/file")


# align the trajectory 
aligner = align.AlignTraj(u, u, select='segid PROA', in_memory=True).run()


# calculate the pairwise distance matrix for atom group selection 1
matrix = diffusionmap.DistanceMatrix(u, select='segid PROA').run()
matrix = matrix.dist_matrix

fig = plt.imshow(matrix, cmap='PuBu', vmin=0, vmax=10)
plt.xlabel('', weight='bold', fontsize=12)

cbar = plt.colorbar(label='RMSD (Å)')
cbar.ax.tick_params(labelsize=16) 
cbar.set_label('RMSD (Å)', fontsize=16)

ax = plt.gca()

ax.set_xticks([])
ax.set_yticks([])

output_path = '/path/to/output/file'
plt.savefig(output_path, format='pdf',bbox_inches="tight")
plt.close()

# calculate the pairwise distance matrix for atom group selection 2
matrix = diffusionmap.DistanceMatrix(u, select='segid PROB').run()
matrix = matrix.dist_matrix

fig = plt.imshow(matrix, cmap='PuBu', vmin=0, vmax=10)
plt.xlabel('', weight='bold', fontsize=12)
plt.ylabel('Sep1-Sep21 complex', weight='bold', fontsize=12)

cbar = plt.colorbar(label='RMSD (Å)')
cbar.ax.tick_params(labelsize=16) 
cbar.set_label('RMSD (Å)', fontsize=16) 

ax = plt.gca()

ax.set_xticks([])
ax.set_yticks([])

output_path = '/path/to/output/file'
plt.savefig(output_path, format='pdf',bbox_inches="tight")
plt.close()
```
